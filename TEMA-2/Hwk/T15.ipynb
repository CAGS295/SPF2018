Para ver si la venta de chips de silicio son independientes del punto del ciclo de negocios en que se encuentre la economía del país se han recogido las ventas semanales de una empresa y datos acerca de la economía del país, y se reportan los siguientes resultados:


Realice la prueba de chi-cuadrado para validar la independencia del punto del ciclo de negocios en que se encuentre la economía.

```python
import numpy as np
a= np.array([[20,7,3],[30,40,30],[20,8,2],[30,5,5]])
b=np.array([30,100,30,40])/200
c=200
```

Para $\alpha$ .05.

```python
d=np.array([*map(lambda x:a.sum(axis=0)*x,b)])
```

```python
from scipy import stats as s
```

```python
print('Xi',((a-d)**2/d).sum())
print(s.chi2.ppf(q = 0.95,df=6))
```

Estimador $\chi^2$ por arriba de límite por lo tanto se rechaza la hipótesis nula y se concluye que las ventas dependen del ciclo de negocio.

```python
s.chisquare(a.flatten(),d.flatten(),)
```

p-value por debajo de $\alpha$ por lo tanto se rechaza la hipótesis nula y se concluye que las ventas dependen del ciclo de negocio.$
