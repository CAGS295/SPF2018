Creamos una estructura que nos permita iterar entre procesos,cada nodo tiene los parámetros que necesita su distribución y la relación a los nodos anteriores. La primera tupla contiene los estimadores media y varianza del proceso según:

$$ \mu = \frac{o+4m+p}{6}$$
$$ \sigma^2 = \frac{(p-o)^2}{6}$$

```python
nodesb=[
 ((1,3,2,1/9),[None]),#"A":0
 ((2,8,4,1),[0]),#"B":1
 ((6,18,10,4),[1]),#"C":2
 ((4,10,6,1),[2]),#"D":3
 ((1,5,4,4/9),[2]),#"E":4
 ((4,10,5,1),[4]),#"F":5
 ((5,11,7,1),[3]),#"G":6
 ((5,17,9,4),[4,6]),#"H":7
 ((3,9,7,1),[2]),#"I":8
 ((3,9,8,1),[5,8]),#"J":9
 ((4,4,4,0),[9]),#"K":10
 ((1,7,5,1),[9]),#"L":11
 ((1,3,2,1/9),[7]),#"M":12
 ((5,9,6,4/9),[10,11]), #"N":13
]
```

Creamos una función que se encarga de simular el tiempo del proceso

```python
def helperb(x):    
    alpha = lambda m,v:((1-m)/v-1/m)*m**2
    beta = lambda m,a:a*(1/m-1)   

    if (x[1]-x[0])==0:
        return x[1]    
    
    muN = (x[2]-x[0])/(x[1]-x[0])
    varN = x[3]/(x[1]-x[0])**2
    alfa=alpha(muN,varN)
    res = st.beta.rvs(a=alfa,b=beta(muN,alfa))                     
    return res*(x[1]-x[0])+x[0]
```

Creamos una función que agregue el tiempo del proceso $n$ al máximo de sus nodos padres.

```python
def aggreg(node,helper,container):     
    return helper(node[0]) + max(([container[i] for i in node[1]]) if node[1][0] else [0])
```

Creamos la función que simula el escenario completo, este crea un contenedor para guardar la suma de los tiempos.

```python
def run(nodes,helper,container):    
#     container = np.zeros(len(nodes))  
    for ix,i in enumerate(nodes):        
        container[ix] =  aggreg(i,helper,container)
#     print(len(container))
    return container[-1] 
```

La función maestra que orquesta las simulaciones.

```python
def sim(nodes,helper,nsims):
    container = np.zeros(len(nodes))  
    return list(map(lambda x: run(nodes,helper,container),range(nsims)))
```

Considerando la simulación con distribuciones beta, se sugiere que la media no es 44 sino 42.14 y la desviación 2.942

```python
Y =sim(nodesb,helperb,10000)
plt.hist(Y,bins=30)
print('Mean',np.mean(Y))
print('SD',np.std(Y))
plt.show()
```

Al cambiar los estimadores $\sigma$ y $\mu$ la probabilidad acumulada $P(T<=d)$ resulta mayor a las pruebas anteriores.

```python
print('P(T<=d)=',reduce(lambda x,y:x+1 if(y<d) else x,Y,0)/10000)
```

>4.Repetir el literal 3 pero en este caso usando una distribución triangular.

Creamos una estructura que nos permita iterar entre procesos,cada nodo tiene los parámetros que necesita su distribución y la relación a los nodos anteriores. La primera tupla tiene el valor optimista,medio y pesimista respectivamente.

```python
nodes=[
 ((1,2,3),[None]),#"A":0
 ((2,3.5,8),[0]),#"B":1
 ((6,9,18),[1]),#"C":2
 ((4,5.5,10),[2]),#"D":3
 ((1,4.5,5),[2]),#"E":4
 ((4,4,10),[4]),#"F":5
 ((5,6.5,11),[3]),#"G":6
 ((5,8,17),[4,6]),#"H":7
 ((3,7.5,9),[2]),#"I":8
 ((3,9,9),[5,8]),#"J":9
 ((4,4,4),[9]),#"K":10
 ((1,5.5,7),[9]),#"L":11
 ((1,2,3),[7]),#"M":12
 ((5,5.5,9),[10,11]), #"N":13
]
```

```python
def helper(x):
#     print(x)
    if (x[2]-x[0])==0:
        return x[1]    
    return st.triang.rvs(loc=x[0],c=(x[1]-x[0])/(x[2]-x[0]),scale=x[2]-x[0])                     
```

Usando una distribución triangular resulta congruente con los resultados previamente obtenidos a excepción de la simulación con distribución beta.

```python
print('P(T<=d)=',reduce(lambda x,y:x+1 if(y<d) else x,sim(nodes,helper,10000),0)/10000)
```

Parece que utilizar una distribución normal cuando el número de variables es grande sin importar la distribución de cada una de las variables cabe considerarse por su facilidad de cálculo y precisión cercana a la simulada considerando la distribución de cada una de las variables; por el teorema de límite central la distribución de la suma de estas se comporta de manera normal. No obstante en el literal 3 obtuvimos resultados diferentes que sugieren que los estimadores $\mu$ y $\sigma$ son diferentes, se pueden haer pruebas de hipótesis para refutar la idea.
