$$\hat \mu = {1\over n}\sum_{i=1}^n x_i,\quad \hat  \sigma^2={1\over n}\sum_{i=1}^n (x_i-\hat \mu)^2$$

**Recuerde que:** La distribuci칩n normal es
$$f(x\mid \mu ,\sigma ^{2})={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}}}$$

$l(x)=(2\pi\sigma^2)^{\frac{-n}{2}}e^\frac{\sum(x_i-\mu)^2}{2\sigma^2} \rightarrow logl(x)=\frac{-n}{2}ln(2\pi\sigma^2)-\frac{\sum(x_i-\mu)^2}{2\sigma^2}$
$$$$
$f_\mu = \frac{-2\sum x_i+2n\mu}{2\sigma^2} \rightarrow f_\mu = 0 \rightarrow \sum X_i -n\mu = \sigma^2$,
$\mu =\frac{ \sum X_i}{n}$
$$$$
$f_\sigma = \frac{-n}2 \frac{4\pi \sigma}{2\pi \sigma^2}+\frac1 2 2 \sigma^{-3} \sum ( x_i -u)^2 \rightarrow \sigma^2= \frac{\sum(x_i-\mu)^2}{n}$
